```python
import json
import csv

with open('branches.json') as json_file:
    data = json.load(json_file)

branches_data = data['branches']
data_file = open('branches.csv', 'w')

csv_writer = csv.writer(data_file)

count = 0

for b in branches_data:
    if count == 0:
        header = b.keys()
        csv_writer.writerow(header)
        count += 1
    csv_writer.writerow(b.values())

data_file.close()
```

```python
# Importing the required libraries
import xml.etree.ElementTree as Xet
import pandas as pd

cols = ["employee_id", "branch_id", "first_name", "last_name", "position"]
rows = []

xmlparse = Xet.parse('employees.xml')
root = xmlparse.getroot()
for i in root:
    employee_id = i.find("employee_id").text
    branch_id = i.find("branch_id").text
    first_name = i.find("first_name").text
    last_name = i.find("last_name").text
    position = i.find("position").text

    rows.append({"employee_id": employee_id,
                 "branch_id": branch_id,
                 "first_name": first_name,
                 "last_name": last_name,
                 "position": position})


df=spark.createDataFrame(data=rows)
df.printSchema()
df.show()

cf=df.toPandas()
cf.to_csv('employees.csv',index=False)
```

```python
empschema = StructType([StructField("branch_id", IntegerType()),
                     StructField("employee_id", IntegerType()),
                     StructField("first_name", StringType()),
                     StructField("last_name", StringType()),
                     StructField("position", StringType())])

emp= spark.read.csv(path='employees.csv',header=True,schema=empschema)
emp.printSchema()
emp.show()
```

```python
import sqlite3

conn = sqlite3.connect('Customers')

cur = conn.cursor()

cur.execute('select * from customers')
l=cur.fetchall()
conn.commit()
conn.close()
print(l)
```

```python
#colm=['customer_id','first_name','last_name','dob']

schema =  StructType([StructField("customer_id", IntegerType()),
                     StructField("first_name", StringType()),
                     StructField("last_name", StringType()),
                     StructField("dob", StringType())])

cus = spark.createDataFrame(data=l,schema=schema)

cus.printSchema()
cus.show()
```

```python
#Problem Statement - 4
#Find the Oldest Customer:
from pyspark.sql.window import Window

#cus.show()

df4 = cus.withColumn('dob_datetype', to_date(from_unixtime(unix_timestamp(col('dob'), 'yyyy-MM-dd')))) \
.withColumn('min', min(col('dob_datetype')).over(Window.orderBy('dob_datetype'))).filter(col('dob_datetype')==col('min')).drop('min')

df4.show()
```

```python
#Problem Statement - 5
#Calculate the Total Interest Earned Across All Accounts:
df5=loans.withColumn('startdate_datetype',to_date(from_unixtime(unix_timestamp(col('start_date'),'yyyy-MM-dd'))))\
.withColumn('enddate_datetype',to_date(from_unixtime(unix_timestamp(col('end_date'),'yyyy-MM-dd'))))\
.withColumn('tenure',months_between('enddate_datetype','startdate_datetype').cast('int'))\
.withColumn('simple_intrest',(col('loan_amount').cast('int')*col('interest_rate').cast('float')*col('tenure'))/100)\
.drop('startdate_datetype','enddate_datetype')

#.join(payment_history,"loan_id","left").sort(col('loan_id').asc())

#df5.show()
sf=payment_history.groupBy('loan_id').count()

df_6=df5.join(sf,'loan_id','left')

df6=df_6.withColumn('tot_intrest_ernd',col('simple_intrest')*col('count'))\
.select('loan_id','tot_intrest_ernd')

df6.show()
```
